```python
def report(df):
    col = []
    d_type = []
    uniques = []
    n_uniques = []
    
    for i in df.columns:
        col.append(i)
        d_type.append(df[i].dtypes)
        uniques.append(df[i].unique()[:5])
        n_uniques.append(df[i].nunique())
    
    return pd.DataFrame({'Column': col, 'd_type': d_type, 'unique_sample': uniques, 'n_uniques': n_uniques})
```

```python
report(Telcochurn)
```

```python
Telcochurn1["Churn Reason"].unique()
array(['Attitude of support person', 'Competitor offered higher download speeds', 'Competitor offered more data', 
       'Dont know', 'Competitor made better offer', 'Attitude of service provider', 'Competitor had better devices',
       'Network reliability', 'Product dissatisfaction', 'Price too high', 'Service dissatisfaction', 
       'Lack of self-service on Website', 'Extra data charges', 'Moved', 'Limited range of services', 'Long distance charges',
       'Lack of affordable download/upload speed', 'Poor expertise of phone support', 'Poor expertise of online support',
       'Deceased'], dtype=object)
def  ChurnR (series):
    if series == "Attitude of support person":
        return 0
    if series == "Competitor offered higher download speeds":
        return 1
    if series == "Competitor offered more data":
        return 2
    if series == 'Dont know':
        return 3
    if series == 'Competitor made better offer':
        return 4
    if series == 'Attitude of service provider':
        return 5
    if series == 'Competitor had better devices':
        return 6
    if series == 'Network reliability':
        return 7
    if series == 'Product dissatisfaction':
        return 8
    if series == 'Price too high':
        return 9
    if series == 'Service dissatisfaction':
        return 10
    if series == 'Lack of self service on Website':
        return 11
    if series == 'Extra data charges':
        return 12
    if series == 'Moved':
        return 13
    if series == 'Limited range of services':
        return 14
    if series == 'Long distance charges':
        return 15
    if series == 'Lack of affordable download/upload speed':
        return 16
    if series == 'Poor expertise of phone support':
        return 17
    if series == 'Poor expertise of online support':
        return 18
    if series == 'Deceased':
        return 19
Telcochurn1["ChurnR"] = Telcochurn1["Churn Reason"].apply(ChurnR)
```

```python
Telcochurn3["Contract"].unique()
array(['Month-to-month', 'Two year', 'One year'], dtype=object)
def ContractR (series):
    if series == "Month-to-month":
        return 0
    if series == "One year":
        return 1
    if series == "Two year":
        return 2
Telcochurn3["ContractR"] = Telcochurn3["Contract"].apply(ContractR)
```

```python
Telcochurn3['Gender'].unique()
```

```python
Telcochurn3["Gender"].unique()
array(['Male', 'Female'], dtype=object)
def GenderR (series):
    if series == "Male":
        return 0
    if series == "Female":
        return 1
Telcochurn3["GenderR"] = Telcochurn3["Gender"].apply(GenderR)
```
