```python
def report(df):
    col = []
    d_type = []
    uniques = []
    n_uniques = []
    
    for i in df.columns:
        col.append(i)
        d_type.append(df[i].dtypes)
        uniques.append(df[i].unique()[:5])
        n_uniques.append(df[i].nunique())
    
    return pd.DataFrame({'Column': col, 'd_type': d_type, 'unique_sample': uniques, 'n_uniques': n_uniques})
```

```python
report(Telcochurn)
```

```python
Telcochurn3["Contract"].unique()
array(['Month-to-month', 'Two year', 'One year'], dtype=object)
def ContractR (series):
    if series == "Month-to-month":
        return 0
    if series == "One year":
        return 1
    if series == "Two year":
        return 2
Telcochurn3["ContractR"] = Telcochurn3["Contract"].apply(ContractR)
```

```python
Telcochurn3.head()
```

```python
Telcochurn3['Gender'].unique()
```

```python
Telcochurn3["Gender"].unique()
array(['Male', 'Female'], dtype=object)
def GenderR (series):
    if series == "Male":
        return 0
    if series == "Female":
        return 1
Telcochurn3["GenderR"] = Telcochurn3["Gender"].apply(GenderR)
```
